# B. Data Cleaning and Preprocessing:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.impute import SimpleImputer
from scipy.stats import ttest_ind
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
wine_data = pd.read_csv('wine.csv')
wine_data.head()
```

```python
wine_data.shape
```

```python
wine_data.info()
```

### • Handle missing data using appropriate imputation strategies.

```python
missing_values = wine_data.isnull().sum()
print(missing_values)

wine_data['country'].fillna(wine_data['country'].mode()[0], inplace=True)
wine_data['province'].fillna(wine_data['province'].mode()[0], inplace=True)
imputer = SimpleImputer(strategy='median')
wine_data['price'] = imputer.fit_transform(wine_data[['price']])
wine_data['designation'].fillna('Unknown', inplace=True)
wine_data['region_1'].fillna('Unknown', inplace=True)
wine_data['region_2'].fillna('Unknown', inplace=True)

print(wine_data.isnull().sum())
```

```python
wine_data.shape
```

### • Detect and address outliers or anomalies in the dataset.

```python
sns.boxplot(x=wine_data['points'])
plt.show()

z_scores = np.abs((wine_data['points'] - wine_data['points'].mean()) / wine_data['points'].std())
wine_data = wine_data[z_scores < 3]

print(wine_data.shape)
```

### • Perform any necessary feature engineering or transformation.

```python
wine_data = wine_data.drop(columns=['Unnamed: 0'], errors='ignore')
wine_data.head()
```

```python
wine_data.shape
```

# E. Hypothesis Testing:

## t-test

```python
points_france = wine_data[wine_data['country'] == 'France']['points']
points_italy = wine_data[wine_data['country'] == 'Italy']['points']
t_stat, p_value = stats.ttest_ind(points_france, points_italy, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("There is a significant difference in the points between wines from France and Italy.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is no significant difference in the points between wines from France and Italy.")

plt.figure(figsize=(10, 6))
sns.histplot(points_france, color='blue', alpha=0.7, label='France')
sns.histplot(points_italy, color='orange', alpha=0.7, label='Italy')
plt.title('Distribution of Points for Wines from France and Italy')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## chi-square

```python
observed_table = pd.crosstab(wine_data['country'], wine_data['variety'])
chi2_stat, p_value, _, _ = stats.chi2_contingency(observed_table)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("There is a significant association between country and wine variety.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is no significant association between country and wine variety.")

plt.figure(figsize=(120, 80))
sns.heatmap(observed_table, annot=True, cmap='Blues')
plt.title('Observed Counts of Wine Varieties by Country')
plt.xlabel('Variety')
plt.ylabel('Country')
plt.show()
```

```python
observed_table = pd.crosstab(wine_data['country'], wine_data['variety'])

top_varieties = observed_table.sum(axis=0).nlargest(10).index
top_countries = observed_table.sum(axis=1).nlargest(10).index

subset_observed_table = observed_table.loc[top_countries, top_varieties]
chi2_stat, p_value, _, _ = stats.chi2_contingency(subset_observed_table)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

alpha = 0.05 
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("There is a significant association between top 10 country and top 10 wine variety.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is no significant association between top 10 country and top 10 wine variety.")

plt.figure(figsize=(16, 10))
sns.heatmap(subset_observed_table, annot=True, cmap='Blues', fmt='d')
plt.title('Observed Counts of Top Varieties by Top Countries')
plt.xlabel('Variety')
plt.ylabel('Country')
plt.show()
```

# F. Advanced Analysis:

#### using  Machine learning 

```python
features = ['price']
target = 'points'
wine_data[features] = wine_data[features].fillna(wine_data[features].mean())

X_train, X_test, y_train, y_test = train_test_split(wine_data[features], wine_data[target], test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2): {r2}')

plt.figure(figsize=(10, 4))
sns.scatterplot(x=X_test['price'], y=y_test, color='blue', label='Actual Points')
sns.lineplot(x=X_test['price'], y=y_pred, color='red', label='Regression Line')
plt.title('Regression Line for Predicting Points based on Price')
plt.xlabel('Price')
plt.ylabel('Points')
plt.legend()
plt.show()
```
